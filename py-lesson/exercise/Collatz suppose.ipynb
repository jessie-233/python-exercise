```python
#Collatz suppose
import random
num = random.randint(1, 10 ** 10)
print(num)
while num != 1:
    if num % 2 == 0:
        num //= 2
    else:
        num = num * 3 + 1
    print("\t",num)
```

```python
for idx in range(3, 16):
    print(idx, end = ': ')
    value = idx
    while value != 1:
        value = value * 3 + 1 if value % 2 else value // 2
        print(value, end = ' ')
    else:
        print()
```
