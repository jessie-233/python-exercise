```python
%matplotlib inline
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
```

```python
time = np.arange(0.0, 2.0, 0.01)
volt = 1 + np.sin(2 * np.pi * time)
```

```python
fig, ax = plt.subplots()
ax.plot(time, volt)
ax.set(xlabel = 'Time(s)', ylabel = 'Voltage(V)', title = 'V-T' )
ax.grid()
plt.show()
```
